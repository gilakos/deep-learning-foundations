# Predictive Theory
If terrible, bad included = Negative
Else if good, excellent = Positive
Else Negative

```python
##Hybrid from example

#import libraries
from collections import Counter
import numpy as np

#create some counters for word counts
pos_count = Counter()
neg_count = Counter()
tot_count = Counter()

#loop through all reviews
for i in range(len(reviews)):
    #check corresponding label
    if (labels[i] == 'POSITIVE'):
        #loop through all words in review
        for word in reviews[i].split(" "):
            pos_count[word] += 1
            tot_count[word] += 1
    else:
        #loop through all words in review
        for word in reviews[i].split(" "):
            neg_count[word] += 1
            tot_count[word] += 1
        
```

```python
#display the most common positive words
pos_count.most_common()
```

```python
#create a counter for ratios
pos_neg_ratios = Counter()

#loop through the terms
for term,cnt in list(tot_count.most_common()):
    #only continue if the term appears more than x times
    if(cnt > 500):
        #define the ratio
        pos_neg_ratio = pos_count[term] / float(neg_count[term]+1)
        #add the ratio to the counter
        pos_neg_ratios[term] = pos_neg_ratio
#loop through the words stored in the counter
for word,ratio in pos_neg_ratios.most_common():
    #if it's positive, apply log
    if(ratio > 1):
        pos_neg_ratios[word] = np.log(ratio)
    #if it's negative apply inverse log
    else:
        pos_neg_ratios[word] = -np.log((1 / (ratio+0.01)))
```

```python
# words most frequently seen in a review with a "POSITIVE" label
pos_neg_ratios.most_common()
```

```python
# words most frequently seen in a review with a "NEGATIVE" label
list(reversed(pos_neg_ratios.most_common()))[0:30]
```
